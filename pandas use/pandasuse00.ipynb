```python
# Data structure : series(1D), Dataframe(2D), Panel(3D)
li=[0,1,0,2,5,1]  
lil=['A','B','C','D','E','F']
marks=pd.Series(li,lil) #(Data, index) if index not given then it will be default start from 0
print(marks)
```

```python
# dataframe (2D)
online=[1,0,1,1]
offline=[0,0,1,2]
lil=['A','B','C','D'] # column name
lig=['Z','X'] # index name
marks=pd.DataFrame([online,offline],lig,lil) # (Data, columns, index)
print(marks)
```
